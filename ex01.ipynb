## 首先利用pd.read_csv()把檔案滙入

我到了https://www.investing.com/  
找出了US Dollar Index Futures和gold 1年的資料下載  
利用panda的read_csv()把資料滙入 index_col把index設定成是Date

```python
import pandas as pd
import numpy as np
gold=pd.read_csv('Gold Futures Historical Data.csv',index_col="Date")
us_index=pd.read_csv('US Dollar Index Futures Historical Data.csv',index_col="Date")
gold.columns=['gold_price','gold_open','gold_high','gold_low','gold_vol.','gold_change%']
us_index.columns=['USDIndex_price',
                  'USDIndex_open',
                  'USDIndex_high',
                  'USDIndex_low',
                  'USDIndex_vol.',
                  'USDIndex_change%']
gold.head()
us_index.head()
```

## 利用concat()把資料合併

```python
import pandas as pd
import numpy as np
gold=pd.read_csv('Gold Futures Historical Data.csv',index_col="Date")
us_index=pd.read_csv('US Dollar Index Futures Historical Data.csv',index_col="Date")
gold.columns=['gold_price','gold_open','gold_high','gold_low','gold_vol.','gold_change%']
us_index.columns=['USDIndex_price',
                  'USDIndex_open',
                  'USDIndex_high',
                  'USDIndex_low',
                  'USDIndex_vol.',
                  'USDIndex_change%']
gold_us=pd.concat([gold,us_index],axis=1)
gold_us.head()
```

## 選擇只看change%

```python
import pandas as pd
import numpy as np
gold=pd.read_csv('Gold Futures Historical Data.csv',index_col="Date")
us_index=pd.read_csv('US Dollar Index Futures Historical Data.csv',index_col="Date")
gold.columns=['gold_price','gold_open','gold_high','gold_low','gold_vol.','gold_change%']
us_index.columns=['USDIndex_price',
                  'USDIndex_open',
                  'USDIndex_high',
                  'USDIndex_low',
                  'USDIndex_vol.',
                  'USDIndex_change%']
gold_us=pd.concat([gold,us_index],axis=1)
gold_us_1=gold_us.dropna(axis=0,how='any')
gold_us_2=gold_us_1.loc[:,['gold_change%','USDIndex_change%']]
gold_us_2.head()
```
