## Selecting columns using .loc[]

```python
import pandas as pd
data=pd.read_csv('msleep_ggplot2.csv')
print(data.loc[:,['name','sleep_total']])
```

## To select a range of columns by name, use the “:” (colon) operator

```python
import pandas as pd
data=pd.read_csv('msleep_ggplot2.csv')
print(data.loc[:,'name':'order'])
```
